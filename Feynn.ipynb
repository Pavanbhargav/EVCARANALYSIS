```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

features = df[["Ev_penetration", "total_Ev_sold", "PCS_Mar_2024", "no_of_ROs_charging"]].dropna()

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Finding the optimal K using the Elbow Method
wcs = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(features_scaled)
    wcs.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(k_range, wcs, marker="o", linestyle="--", color="b")
plt.xlabel("Number of Clusters - K")
plt.ylabel("Within-Cluster Sum of Square")
plt.title("Elbow Method for Optimal K")
plt.xticks(k_range)
plt.show()

optimal_k = np.diff(wcs, 2).argmin() + 2
print(f"Optimal K detected: {optimal_k}")
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(features_scaled)
    silhouette_scores.append(silhouette_score(features_scaled, cluster_labels))

plt.figure(figsize=(8, 5))
plt.plot(range(2, 10), silhouette_scores, marker="o", linestyle="--", color="purple")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("Optimal Cluster Count")
plt.show()
```

```python
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df["Segment"] = kmeans.fit_predict(features_scaled)

print(df["Segment"].value_counts())
```

```python
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="Segment", y="total_Ev_sold", palette="coolwarm", ci=None)
plt.xlabel("Cluster Segment")
plt.ylabel("Total EVs Sold")
plt.title(f"EV Sales per Cluster Segment (K={optimal_k})")
plt.show()
```

```python
import plotly.express as px

fig = px.scatter_3d(df, x="total_Ev_sold", y="Ev_penetration", z="PCS_Mar_2024",
                    color=df["Segment"].astype(str),
                    size="no_of_ROs_charging",
                    hover_data=["state_ut"], title=f"EV Market Segmentation (K={optimal_k})")

fig.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_results = pca.fit_transform(features_scaled)
df["PCA_1"] = pca_results[:, 0]
df["PCA_2"] = pca_results[:, 1]

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="PCA_1", y="PCA_2", hue="Segment", palette="viridis", edgecolor="black", s=100)
plt.title("EV Market Segments")
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

linkage_matrix = linkage(features_scaled, method="ward")

plt.figure(figsize=(12, 6))
dendrogram(linkage_matrix, labels=df["state_ut"].values, leaf_rotation=90, leaf_font_size=10)
plt.title("EV Market Segmentation")
plt.xlabel("States")
plt.ylabel("Distance")
plt.show()
```
